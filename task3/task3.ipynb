# import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Line Plots with matplotlip

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun","Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales = [1200, 1350, 1500, 1700, 2000, 2300, 2600, 2500, 2200, 2100, 1900, 1800]

plt.plot(months,sales,marker="*",linestyle="--",color="purple")
plt.title("Monthly Sales Performance")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.style.use("ggplot")
```

# Line plots with seaborn

```python
sales_df = pd.DataFrame({"Months":months,"Sales":sales})
sns.lineplot(data=sales_df,x="Months",y="Sales",marker="o",linestyle=":",color="green")
plt.title("Monthly Sales Performance")
plt.grid(True)
sns.set_style("dark")
```

# Scatter plot

```python
iris_df=sns.load_dataset("iris")
print(iris_df.head())
```

```python
sns.scatterplot(data=iris_df,x="sepal_length",y="sepal_width",hue="species")
sns.set_style("whitegrid")
plt.title("Sepal Length vs Width by Species")
plt.show()
```

# Histograms plots

```python
tips_df=sns.load_dataset("tips")
print(tips_df.head())
```

```python
sns.histplot(data=tips_df,x="tip",bins=10,kde=True,color="blue")
plt.title("Distribution of Tips")
plt.show()
```

# KDE plots

```python
sns.kdeplot(data = tips_df,x="total_bill",hue="day",fill=True)
plt.title("Distribution of total bill by days")
plt.show()
```

# ECDF plots

```python
sns.ecdfplot(data=tips_df,x="tip",hue="sex")
plt.title("Cumulative Tip Distribution by sex")
plt.show()
```

# Bar plots

```python
sns.barplot(data=tips_df,x="sex",y="tip",hue="day")
plt.title("Cumulative Tip Distribution by sex")
plt.show()
```

# Box plots

```python
sns.boxplot(data=tips_df,x="sex",y="tip",hue="day")
plt.title("Cumulative Tip Distribution by sex")
sns.set_style("darkgrid")
plt.show()
```

# Violin plots

```python
sns.violinplot(data=tips_df,x="sex",y="tip",hue="day")
plt.title("Cumulative Tip Distribution by sex")
sns.set_style("darkgrid")
plt.show()
```

# Heatmaps 

```python
flight_df=sns.load_dataset("flights").pivot(index="month",columns="year",values="passengers")
flight_df.head()
```

```python
sns.heatmap(data=flight_df,fmt="d",annot=True)
plt.title("Monthly Flight Passengers")
plt.show()
```

# Subplots

```python
fig , ax = plt.subplots(2,2,figsize=(8,8))

sns.kdeplot(data = tips_df,x="total_bill",hue="day",fill=True,ax=ax[0,0])
ax[0,0].set_title("Distribution of total bill by days")


sns.boxplot(data=tips_df,x="sex",y="tip",hue="day",ax=ax[0,1])
ax[0,1].set_title("Cumulative Tip Distribution by sex")


sns.boxplot(data=tips_df,x="sex",y="tip",hue="day",ax=ax[1,0])
ax[1,0].set_title("Cumulative Tip Distribution by sex")


sns.violinplot(data=tips_df,x="sex",y="tip",hue="day",ax=ax[1,1])
ax[1,1].set_title("Cumulative Tip Distribution by sex")

plt.tight_layout()
plt.show()





```
