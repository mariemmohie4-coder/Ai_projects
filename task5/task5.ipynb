<h2 style = "color:blue"> 1- Spliting Data </h2>

```python
import numpy as np
import pandas as pd
data = pd.read_csv("California_Houses.csv")
#print(data.info())
N=data.shape[0]     #20640 rows
indices=np.random.permutation(N)
training=data.iloc[indices[:int(0.7*N)]]
validation=data.iloc[indices[int(0.7*N):int(0.85*N)]]
test=data.iloc[indices[int(0.85*N):]]

x_training=training.drop("Median_House_Value",axis=1).values
y_training=training["Median_House_Value"].values

x_validation=validation.drop("Median_House_Value",axis=1).values
y_validation=validation["Median_House_Value"].values

x_test=test.drop("Median_House_Value",axis=1).values
y_test=test["Median_House_Value"].values
```

<h2 style = "color:blue"> 2- Linear Regression from scratch </h2>

<h2 style = "color:red"> 2.1- Computing weights </h2>

<h2 style = "color:green"> Feature Scaling </h2>

```python
x_mean = x_training.mean(axis=0)
x_std = x_training.std(axis=0)
x_std[x_std == 0] = 1


x_training_scaled = (x_training - x_mean)/x_std
x_validation_scaled = (x_validation - x_mean)/x_std
x_test_scaled = (x_test - x_mean)/x_std

x_training_b_scaled=np.c_[np.ones((len(x_training_scaled),1)),x_training_scaled]
x_validation_b_scaled=np.c_[np.ones((len(x_validation_scaled),1)),x_validation_scaled]
x_test_b_scaled=np.c_[np.ones((len(x_test_scaled),1)),x_test_scaled]

```

```python
y_mean = y_training.mean()
y_std = y_training.std()

y_training_scaled = (y_training - y_mean)/y_std
y_validation_scaled = (y_validation - y_mean)/y_std
y_test_scaled = (y_test - y_mean)/y_std
```

```python
#Training
wn= (np.linalg.inv((x_training_b_scaled.T)@x_training_b_scaled))@(x_training_b_scaled.T)@y_training_scaled
h1=x_training_b_scaled @ wn

#Validation
h2=x_validation_b_scaled @ wn

#Testing
h3 = x_test_b_scaled @ wn
```

<h2 style = "color:green"> MSE & MAE </h2>

```python
#Training
mse1=(h1-y_training_scaled)**2
j1=0.5*np.mean(mse1)
print(f"MSE for Training = {j1}")

mae1=np.abs(h1-y_training_scaled)
j2=0.5*np.mean(mae1)
print(f"MAE for Training = {j2}")

#Validation
mse2=(h2-y_validation_scaled)**2
j3=0.5*np.mean(mse2)
print(f"MSE for Validation normal = {j3}")

mae2=np.abs(h2-y_validation_scaled)
j4=0.5*np.mean(mae2)
print(f"MAE for Validation normal = {j4}")

#Testing
mse3=(h3-y_test_scaled)**2
j5=0.5*np.mean(mse3)
print(f"MSE for Testing normal = {j5}")

mae3=np.abs(h3-y_test_scaled)
j6=0.5*np.mean(mae2)
print(f"MAE for Testing normal = {j6}")
```

<h2 style = "color:red"> 2.2- Gradient Descent </h2>

<h2 style = "color:green"> Weights & Learning Rate </h2>

```python
#Training
wg = np.zeros(x_training_b_scaled.shape[1])
interations=1000
alpha=0.1
m = len(x_training_b_scaled)
for i in range(interations) :
    hg = x_training_b_scaled @ wg
    grad = (1/m)*((x_training_b_scaled.T)@(hg-y_training_scaled))
    wg = wg - alpha*grad

    if i % 100 == 0 :
        losses = 0.5 * np.mean((hg-y_training_scaled)**2)
        print(f"Iinterations = {i} , MSE = {losses} ")
print(f"MSE for Training Gradient = {0.5 * np.mean((x_training_b_scaled @ wg - y_training_scaled)**2)}")
print(f"MAE for Training Gradient = {0.5 * np.mean(np.abs(x_training_b_scaled @ wg - y_training_scaled))}")
```

```python
#Validation
hgv = x_validation_b_scaled @ wg
mse_v=(hgv-y_validation_scaled)**2
jv1=0.5*np.mean(mse_v)
print(f"MSE for Validation Gradient = {jv1}")

mae_v=np.abs(hgv-y_validation_scaled)
jv2=0.5*np.mean(mae_v)
print(f"MAE for Validation Gradient = {jv2}")
```

```python
#Testing
hgt = x_test_b_scaled @ wg
mse_t=(hgt-y_test_scaled)**2
jt1=0.5*np.mean(mse_t)
print(f"MSE for Testing Gradient = {jt1}")

mae_t=np.abs(hgt-y_test_scaled)
jt2=0.5*np.mean(mae_t)
print(f"MAE for Testing Gradient = {jt2}")
```

<h2 style = "color:blue"> 3- Linear Regression using Scikit-learn
 </h2>

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , mean_absolute_error

model = LinearRegression()
model.fit(x_training_scaled, y_training_scaled)

w_s=model.coef_
alpha_s=model.intercept_
print(f"Weights = {w_s}")
print(f"Bias = {alpha_s}")
```

```python
#Training
y_pred1 = model.predict(x_training_scaled)
mse_s1 = mean_squared_error(y_training_scaled, y_pred1)
mae_s1 = mean_absolute_error(y_training_scaled, y_pred1)

print(f"MSE for Training Scikit Learn = {0.5 * mse_s1}")
print(f"MAE for Training Scikit Learn = {0.5*mae_s1}")
```

```python
#Validation
y_pred = model.predict(x_validation_scaled)
mse_s = mean_squared_error(y_validation_scaled, y_pred)
mae_s = mean_absolute_error(y_validation_scaled, y_pred)

print(f"MSE for Validation Scikit Learn = {0.5 * mse_s}")
print(f"MAE for Validation Scikit Learn = {0.5*mae_s}")
```

```python
#Testing
y_pred_t = model.predict(x_test_scaled)
mse_s_t = mean_squared_error(y_test_scaled, y_pred_t)
mae_s_t = mean_absolute_error(y_test_scaled, y_pred_t)

print(f"MSE for Testing Scikit Learn = {0.5 * mse_s_t}")
print(f"MAE for Testing Scikit Learn = {0.5*mae_s_t}")
```

<h2 style = "color:purple"> 4- Finally, Comparissson & comments </h2>

<h3 style="color: #2e64fe; border-bottom: 2px solid #2e64fe; padding-bottom: 5px;"> Comparison: Normal Equation vs. Gradient Descent</h3>

<div style="background-color: #fcfcfc; padding: 15px; border-radius: 8px; border: 1px solid #e0e0e0;">


<table style="width:100%; border-collapse: collapse; margin: 10px 0;">
  <tr style="background-color: #f2f2f2;">
    <th style="padding: 10px; border: 1px solid #ddd;">Feature</th>
    <th style="padding: 10px; border: 1px solid #ddd;">Normal Equation (Closed-form)</th>
    <th style="padding: 10px; border: 1px solid #ddd;">Gradient Descent (Iterative)</th>
  </tr>
  <tr>
    <td style="padding: 10px; border: 1px solid #ddd; font-weight: bold;">Complexity</td>
    <td style="padding: 10px; border: 1px solid #ddd;">Directly computes the solution using matrix inversion.</td>
    <td style="padding: 10px; border: 1px solid #ddd;">Requires multiple iterations to converge.</td>
  </tr>
  <tr>
    <td style="padding: 10px; border: 1px solid #ddd; font-weight: bold;">Hyperparameters</td>
    <td style="padding: 10px; border: 1px solid #ddd;">None (No alpha or iterations to choose).</td>
    <td style="padding: 10px; border: 1px solid #ddd;">Needs careful tuning of Learning Rate (alpha).</td>
  </tr>
  <tr>
    <td style="padding: 10px; border: 1px solid #ddd; font-weight: bold;">Scaling</td>
    <td style="padding: 10px; border: 1px solid #ddd;">Works fine without feature scaling.</td>
    <td style="padding: 10px; border: 1px solid #ddd;">Feature scaling is <b>essential</b> for performance.</td>
  </tr>
  <tr>
    <td style="padding: 10px; border: 1px solid #ddd; font-weight: bold;">Scalability</td>
    <td style="padding: 10px; border: 1px solid #ddd;">Slow if the number of features is very large.</td>
    <td style="padding: 10px; border: 1px solid #ddd;">Scales very well with large datasets and many features.</td>
  </tr>
</table>

<p style="margin-top: 15px;">
<strong> Conclusion:</strong><br>
For this dataset, the <b>Normal Equation</b> was the most efficient as it gave the exact answer instantly. However, <b>Gradient Descent</b> is more flexible and is the method used in deep learning and larger datasets where matrix inversion is too expensive.
</p>
